## Slice

Supongamos que despues de analizar nuestro DataFrame llegamos a la conclusión de que solo vamos a trabajar con una porcion de toda la información (ya sea por la cantidad de filas o columnas).

Para ello entonces debemos hacer lo que se llama slice, que sería realizar un corte en nuestros datos para extraer solo la porcion que nos sirven para trabajar de ahora en adelante.

Al igual que con el indexado, tanto Numpy como Pandas ofrecen metodos para este fin.

### Numpy

```python
# Supongamos que queremos trabajar con todas las variables, pero solo con las primeras 15 observaciones (cortamos por filas)
df_table_one[:16]
```

```python
# Supongamos que queremos trabajar con todas las variables, pero solo con observaciones que van desde la 10 a la 20 (cortamos por filas)
df_table_one[9:20]
```

```python
# Supongamos que queremos trabajar con todas las filas, pero solo con la variable Age(Cortamos por columnas)
df_table_one['Age']

#Lo puedo convertir en un nuevo DataFrame haciendo lo siguiente
#pd.DataFrame(df_table_one['Age']) # sacar el comentario para ver el resultado
```

### Merge

Este metodo sirve para unir DataFrames utilizando una clave entre ambas bases.

```python
#Supongamos que queremos unir ambas tablas, mediante la columna PassengerId y que solo haga el merge de los valores que encuentre en la tabla de la izquierda
pd.merge(df_table_one,df_table_two,how="left",on="PassengerId") #recomiendo practicar cambiando los valores del parametro how="" para ver como se modifica el nuevo DataFrame
```

Es importante remarcar que el parametro **how=""** es muy importane, ya que determina como se realizará este merge.

Las 4 opciones posibles del parametro son:
* **inner**: tomando como referencia la clave, unira solo los valores que encuentre en ambas tablas.
* **full**: tomando como referencia la clave, unira todos los valores y los que no encuentre los pondrá como NaN.
* **left**: tomando como referencia la clave, solo unirá los valores que encuentre en la tabla de la izquierda.
* **right**: tomando como referencia la clave, solo unirá los valores que encuentre en la tabla de la derecha.

### Concat

Este metodo nos permite agregas nuevas instancias a nuestro DataSet original.

Considero importante remarcar que al momento de agregar nuevas instancias a nuestro DataSet ambos deben tener la misma estructura (las columnas deben ser las miamas).

Antes de comenzar (y para cumplir con la condicion mencionada previamente), voy a diviri el df_table_one en doa para realizar el concat.

```python
#divido el df original en dos partes aleatorias
df_table_one_a = df_table_one[:500]
df_table_one_b = df_table_one[501:]
```

```python
df_table_one_a.head(2)
```

```python
df_table_one_b.head(2)
```

Controlamos que ambas tablas tengna las mismas columnas para hacer el concat

```python
#Vamos a ver que el DataFrame resultante tiene las mismas dimensiones que el original
pd.concat([df_table_one_a,df_table_one_b], axis=0)
```
